```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
import os 
from tqdm import tqdm
```

```python
path = "/PBL/PBL6/train"
sizes = set()
i=0
for img in os.listdir(path):
    try:
        img_file = cv2.imread(path + '/' + img )
        # sizes.add((img.shape[0],img.shape[1]))
        plt.subplot(2,1,1)
        plt.plot(img_file.shape[0],img_file.shape[1],'.')
        i+=1
    except Exception as e:
        print(e)
```

```python
x = np.array(['200-400X500','400-600X500','600-800X500','800-1000X500'])
y = np.array([0,0,0,0])
path = "/PBL/PBL6/train"
for img in os.listdir(path):
    try:
        img_file = cv2.imread(path + '/' + img )
        if(img_file.shape[0] >= 800):
            y[3] = y[3] + 1
        elif(img_file.shape[0] >= 600):
            y[2] = y[2] + 1
        elif(img_file.shape[0] >= 400):
            y[1] = y[1] + 1
        else:
            y[0]= y[0] + 1
    except Exception as e:
        print(e)
plt.bar(x,y)
plt.show()
```
